```python
import pandas as pd
import numpy as np
import statsmodels.api as stn
```

```python
cars = pd.read_csv("cars.csv")
```

```python
cars.head()
```

```python
cars.describe()
```

```python
cars.info()
```

```python
x=cars.drop(['MPG','Car','Origin'],axis=1)
y=cars.MPG
```

```python
x
```

```python
y
```

```python
x=stn.add_constant(x)
```

```python
model=stn.OLS(y,x).fit()
```

```python
model.params
```

```python
model.summary()
```
